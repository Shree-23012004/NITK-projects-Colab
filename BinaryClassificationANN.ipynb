```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import matplotlib as img
from sklearn.model_selection import train_test_split
```

```python
dataset=pd.read_csv('/content/Social_Network_Ads.csv')
print(dataset.head())
print('')
print(dataset.shape)
```

```python
x1=np.array(dataset.iloc[:,2:4])
y=np.array(dataset.iloc[:,4])
print(x1.shape)
print(y.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x1=sc.fit_transform(x1)
# print(x1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x1,y,test_size=0.25)
print(x_train.shape)
print(x_test.shape)
```

```python
def initialize_params(x_train):
  w1 = np.random.randn (5, x_train.shape[1])
  w2 = np.random.randn (1, 5)
  b1 = np.random.randn (5, 1)
  b2 = np.random.randn (1, 1)
  return w1, w2, b1, b2
def sigmoid_act(x):
  s = 1/(1+np.exp(-x))
  return s
def forward_path(x, w1, b1,w2, b2):
#
#
  z1 = np.reshape(np.matmul (w1, x.T), b1. shape)+b1
  print(z1. shape)
  A1 =sigmoid_act (z1)
  print (A1. shape)
  z2 = np.matmul (w2, A1)+b2
  print (z2. shape)
  A2 =sigmoid_act(z2)
  print(A2. shape)
  return A2, A1

def compute_cost (Y_pred, Y_true):

  cost =(Y_true* np.log(Y_pred+(10**-7)))+((1-Y_true) *np.log(1-Y_pred+(10**-7)))
  return -cost

def backward_path (A2, A1, w2, Y,x):
##Output Layer
# dA2 = (A2-Y)/(A2*(1-A2))
  dz2 = (A2-Y)
  dw2= np.matmul (dz2, A1.T)
  db2= dz2.mean (axis=1)
## First Layer
  dA1 =np.matmul (w2.T, dz2)
  dz1 = np.multiply (dA1, (A1*(1-A1)))
  dw1 = dz1*x
  db1 = dz1.mean (axis=1)
  return dw2, dw1, db2, db1

def update_params (w, b, dw, db, learning_rate):
  w = w - learning_rate * dw
  db = np.reshape(db, b.shape)
  b = b - learning_rate * db
  return w,b
```

```python
num_iters = 500
learning_rate = 0.001
his1 = []
w1, w2, b1, b2 = initialize_params (x_train)
for i in range (num_iters):
  cost = 0
  for j in range(len(x_train)):
    A2, A1 = forward_path(x_train [j], w1, b1,w2, b2)
    y_pred = A2
    cost += compute_cost (y_pred, y_train [j])
    dw2, dw1, db2, db1 = backward_path (A2, A1, w2, y_train [j],x_train[j])
    w2, b2 = update_params (w2, b2, dw2, db2, learning_rate)
    w1, b1 = update_params (w1, b1, dw1, db1, learning_rate)
  his1.append(cost/len(x_train))
```

```python
his = np.reshape(np.array (his1), (len (his1)))
plt.plot(his)
plt.show()
```

```python
y_pred =np.zeros(y_test.shape)
for i in range(len(x_test)):
  A2, A1 = forward_path(x_test[i], w1, b1, w2, b2)
  A2 =np.where (A2>0.5,1,0)
  y_pred[i] = A2
print("predicted")
print(y_pred)
print("original")
print(y_test)
```

```python
print('No.of wrong predictions = ', sum(y_pred!=y_test))
```

```python
TP=sum(np.logical_and(y_pred==1,y_test==1))
TN=sum(np.logical_and(y_pred==0,y_test==0))
FP=sum(np.logical_and(y_pred==1,y_test==0))
FN=sum(np.logical_and(y_pred==0,y_test==1))
print('TP = ',TP)
print('TN = ',TN)
print('FP = ',FP)
print('FN = ',FN)
```

```python
accuracy=(TP+TN)/(TP+TN+FP+FN)
print('Accuracy = ',accuracy)
precision=TP/(TP+FP)
print('Precision = ',precision)
recall=TP/(TP+FN)
print('Recall = ',recall)
f1_score=2*precision*recall/(precision+recall)
print('F1 Score = ',f1_score)
```

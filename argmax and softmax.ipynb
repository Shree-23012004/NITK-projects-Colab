```python
import numpy as  np
x=np.array([[0.1,0.5],
            [1.1,2.3],
            [-1.1,-2.3],
            [-1.5,-2.5]])
w=np.array([[0.1,.2,0.3],
            [0.1,0.2,0.3]])

bias=np.array([[0.01,0.1,0.1]])

print(x)
print(w)
print(bias)
```

```python
def net_input(x,w,b):
  return(x.dot(w)+b)
net_in=net_input(x,w,bias)
print(net_in)
```

```python
def softmax(net_in):
  return(np.exp(net_in)/np.sum(np.exp(net_in)))
softmax_out=softmax(net_in)
print(softmax_out)
```

```python
def to_classlable(z):
  return(z.argmax(axis=1))
classlable=to_classlable(softmax_out)
print(classlable)
```

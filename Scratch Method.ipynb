# **Sequential Approach**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model_ann=Sequential()
model_ann.add(Dense(16,input_dim=3,activation='tanh'))
model_ann.add(Dense(32,activation='tanh'))
model_ann.add(Dense(1,activation='linear'))
model_ann.summary()
```

```python
def initialize_params(x):
  w1 = np.random.randn(5, x.shape[1])
  w2 = np.random.randn(1, 5)
  b1 = np.random.randn(5, 1)
  b2 = np.random.randn(1, 1)
  return w1,w2,b1,b2

def sigmoid_act(x):
  s = 1/(1+np.exp(-x))
  return s

def forward_path(X_train,w1,b1,w2,b2):
#Hidden layer
  z1 = np.matmul(w1, X_train.T)+b1
  A1 = sigmoid_act(z1) ### output of hidden layer
#output layer
  z2 = np.matmul(w2, A1)+b2
  A2 = z2 ### Linear actication function at outputv node
  return A2,A1

def compute_cost(Y_pred, Y_train):
  cost = np.mean((Y_pred - Y_train)**2)
  return cost

def backward_path(A2,A1,w2,Y_train,X_train):
## Output Lalyer
  dA2 = (2*(A2-Y_train))/len(Y_train) ## gradient of Loss with respect to A2
  dz2 = dA2 ## gradient of loss with respect to z2
  dw2 = np.matmul(dz2,A1.T) ## gradient of loss with respect to w2
  db2 = dz2.sum(axis=1) ## gradient of loss with respect to b2
## First layer
  dA1 = np.matmul(w2.T,dz2) ## gradient of loss with respect to A1
  dz1 = np.multiply(dA1,A1*(1-A1)) ## gradient of loss with respect to z1
  dw1 = np.matmul(dz1,X_train) ## gradient of loss with respect to w1
  db1 = dz1.sum(axis=1) ## gradient of loss with respect to b1
  return dw2,dw1,db2,db1

def update_params(w,b,dw,db,learning_rate):
  w = w - learning_rate * dw
  db = np.reshape(db,b.shape)
  b = b - learning_rate * db
  return w,b

```

```python
num_iters = 500
learning_rate = 0.001
his1 = []
w1,w2,b1,b2 = initialize_params(x_train)
for i in range(num_iters):
  A2, A1 = forward_path(x_train,w1,b1,w2,b2)
  y_pred = A2
  cost = compute_cost(y_pred, y_train)
  dw2,dw1,db2,db1 = backward_path(A2,A1,w2,y_train,x_train)
  w2,b2 = update_params(w2,b2,dw2,db2,learning_rate)
  w1,b1 = update_params(w1,b1,dw1,db1,learning_rate)
  his1.append(cost)
# print('Cost at iteration ' + str(i+1) + ' = ' + str(cost) + '\n')
```

```python
from tensorflow.keras.layers import Input, Dense # Import Input and Dense from tensorflow.keras.layers
from tensorflow.keras.models import Model
inp=Input(shape=(3,))
x1=Dense(16,activation='tanh')(inp)
x2=Dense(32,activation='tanh')(x1)
out=Dense(1,activation='tanh')(x1)
model_ann2=Model(inputs=inp,outputs=out)
model_ann2.summary()
```

```python
model_ann2.compile(loss='mean_squared_error',optimizer='SGD',metrics=['mean_absolute_error'])
history_ann=model_ann2.fit(x_train,y_train,epochs=100,batch_size=2,validation_data=(x_test,y_test),verbose=1)
```

```python
#plot model accuracy data
print(history_ann.history.keys())

f,ax=plt.subplots()
ax.plot(history_ann.history['mean_absolute_error'])
ax.plot(history_ann.history['val_mean_absolute_error'],'--')
ax.legend(['train','test'])
ax.set_title('Model Accuracy/validation MSE per epoch')
ax.set_ylabel('Accuracy/MAE')
ax.set_xlabel('Epoch')
```

```python
#plot model loss on training
f,ax=plt.subplots()
ax.plot(history_ann.history['loss'])
ax.plot(history_ann.history['val_loss'],'--')
ax.legend(['train','test'])
ax.set_title('Model Loss/validation Loss per epoch')
ax.set_ylabel('Loss')
ax.set_xlabel('Epoch')
```

```python
#model evaluation
score=model_ann2.evaluate(x_test,y_test)
print('test loss=',score[0])
print('test MAE',score[1])
```

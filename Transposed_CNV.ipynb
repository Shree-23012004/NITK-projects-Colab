```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation
import cv2
```

# **Example 1**

# **Step 1 3X3 kernal**

```python
inp=np.array(([4,5,8,7],[1,8,8,8],[3,6,6,4],[6,5,7,5]))
print(inp.shape)
inp
```

# **Step 2 Kernal filter**

```python
k=np.array(([1,4,-1],[1,3,3],[1,0,-2]))
print(k.shape)
k
```

# **Step 3 creating conv Matrix**

```python
w,h=inp.shape
kw,kh=k.shape
s=1
w_o=int((w-kw)/s)+1
h_o=int((h-kh)/s)+1

ext_mat=np.zeros((w*h))
for i in range(kw):
  ext_mat[i*h:i*h+kh]=k[i]
print(ext_mat)
print(ext_mat.shape)
```

```python
ext_ker=[]
for i in range(w_o):
  for j in range(h_o):
    ext_mat1=np.roll(ext_mat,h*i*s+j*s)
    ext_ker.append(ext_mat1)
ext_ker=np.array(ext_ker)
print('Conv Matrix')
print(ext_ker.shape)
print(ext_ker)
```

# **Step 4 Conv feature map**

```python
#converting inp image to column vector
inp=inp.flatten()

#convolution
out=np.matmul(ext_ker,inp)
out=np.reshape(out,(w_o,h_o))
print('Conv Feature Map')
print(out)
print(out.shape)
```

# **Step 5  unsampling using transposed conv**

```python
w,h=out.shape
wk,hk=k.shape
s=1
w_o=int(((w-1)*s)+wk)
h_o=int(((h-1)*s)+hk)
inp=out.flatten()

out=np.matmul(ext_ker.T,inp)
out=np.reshape(out,(w_o,h_o))
print('Unsampling')
print(out)
print(out.shape)
```

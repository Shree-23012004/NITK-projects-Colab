```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```python
#step 1 input matrix or image
inp=np.array(([4,5,8,7],[1,8,8,8],[3,6,6,4],[6,5,7,8]))
print(inp.shape)
inp
```

```python
#step 2 kernal / filter
kernal=np.array(([1,0,-1],[2,0,-2],[1,0,-1]))
print(kernal.shape)
kernal

```

```python
#step 3 create a convolution matrix from 3x3 matrix
w,h=inp.shape
w_k,h_k=kernal.shape
s=1
w_o=int(((w-w_k)/s)+1)
h_o=int(((h-h_k)/s)+1)
print(w_o,h_o)

ext_mat=np.zeros((w*h))
print(ext_mat)

for i in range(w_k):
  ext_mat[i*h:i*h+h_k]=kernal[i]
print(ext_mat)
print(ext_mat.shape)
```

```python
ext_ker=[]
for i in range(w_o):
  for j in range(h_o):
    ext_mat1=np.roll(ext_mat,h*i*s+j*s)
    ext_ker.append(ext_mat1)
ext_ker=np.array(ext_ker)
print("convolution matrix ")
print(ext_ker)
print(ext_ker.shape)
```

```python
#step 4 conv feature map
inp=inp.flatten()
print(inp)

out1=np.matmul(ext_ker,inp)
print(out1)
print('convoluted feature map')
out2=np.reshape(out1,(w_o,h_o))
print(out2)
```

```python
#step 5 max pooling
def maxpool(img,factor):
  ds_img=np.full((img.shape[0]//factor, img.shape[1]//factor),float('inf'),dtype=img.dtype)
  np.maximum.at(ds_img,(np.arange(img.shape[0])[:,None]//factor,np.arange(img.shape[1])//factor),img)
  return ds_img
factor=2
max_out=maxpool(out2,2)
print(max_out)
```
